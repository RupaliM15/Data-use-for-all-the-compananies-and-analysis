# Change in stock price over time

```python
plt.figure(figsize=(15,10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list, 1):
    plt.subplot(2,2,i)
    company['Close'].plot()
    plt.ylabel('Close')
    plt.title(f"Stock Prices of {company_name[i - 1]}")

plt.tight_layout()
```

The Closing price is the last price at which the stock is traded the regular trading day. A stock's closing price is the standard benchmark used by investors to track its performance over time.
Now that we've seen the visulization for the volume traded each day.

# Change in a stock's volume being traded,over time

```python
plt.figure(figsize=(15,10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list, 1):
    plt.subplot(2,2,i)
    company['Volume'].plot()
    plt.ylabel('Volume')
    plt.title(f"Stock Prices of {company_name[i - 1]}")

plt.tight_layout()
```

Volume is the amount of an asset or security that changes hands over some period of time, often over the course of a daow. For instance, the stock trading volume would refer to the number of shares of security traded between its daily open and close. Trading volume, and changes to volume over the course of time, are important inputs for technical traders.
Now that we've seen the visulization for the volume traded each day.


# Moving Averages of various stocks

```python
ma_day = [50,200]

for ma in ma_day:
    for company in company_list:
        column_name = f"MA for {ma} days"
        company[column_name] = company['Close'].rolling(ma).mean()
        

fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

AAPL[['Close', 'MA for 50 days', 'MA for 200 days']].plot(ax=axes[0,0])
axes[0,0].set_title('APPLE')

GOOG[['Close', 'MA for 50 days', 'MA for 200 days']].plot(ax=axes[0,1])
axes[0,1].set_title('GOOGLE')

MSFT[['Close', 'MA for 50 days', 'MA for 200 days']].plot(ax=axes[1,0])
axes[1,0].set_title('MICROSOFT')

AMZN[['Close', 'MA for 50 days', 'MA for 200 days']].plot(ax=axes[1,1])
axes[1,1].set_title('AMAZON')

fig.tight_layout()
```

The moving average (MA) is a simple technical analysis tool that smooths out price data by creating a constantly updated average price. The average is taken over a specific period of time, like 50 days and 200 days or any time period
the trader chooses. We see in the graph that the best values to measure the moving averages.

# Daily Return average of stocks

```python
for company in company_list:
    company['Daily Return'] = company['Close'].pct_change()

# Then we'll plot the daily return percentage
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

AAPL['Daily Return'].plot(ax=axes[0,0],marker='o', color='red')
axes[0,0].set_title('APPLE')

GOOG['Daily Return'].plot(ax=axes[0,1],marker='o', color='brown')
axes[0,1].set_title('GOOGLE')

MSFT['Daily Return'].plot(ax=axes[1,0],marker='o', color='orange')
axes[1,0].set_title('MICROSOFT')

AMZN['Daily Return'].plot(ax=axes[1,1],marker='o', color='Coral')
axes[1,1].set_title('AMAZON')

fig.tight_layout()
```

Now that we've done some baseline analysis, let's go ahead and dive a little deeper. We're now going to analyze the risk of the stock. In order to do so we'll need to take a closer look at the daily changes of the stock, and not just its absolute value. 

# Visualize trend frequency through a Pie chart

```python
amzn_vol_avg = round(AMZN["Volume"].mean())
aapl_vol_avg = round(AAPL["Volume"].mean())
goog_vol_avg = round(GOOG["Volume"].mean())
msft_vol_avg = round(MSFT["Volume"].mean())

vol_avg = np.array([amzn_vol_avg, aapl_vol_avg, goog_vol_avg, msft_vol_avg])
vol_avg


mylabels = ["Amazon","Apple","Google","Microsoft"]
myexplode = [0.2,0,0,0]  # To create an exploding wedge

plt.pie(vol_avg, labels = mylabels, explode = myexplode, shadow = True, autopct="%.2f")
plt.show()
```

From the pie chart, we find that Apple has the highest averages trade volume, followed by Amzon, Microsoft, Google
